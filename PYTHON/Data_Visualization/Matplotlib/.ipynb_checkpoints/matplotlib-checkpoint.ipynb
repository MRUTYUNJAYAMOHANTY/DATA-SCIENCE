# Functional Plot

```python
plt.plot(x,y,'r') # 'r' is the color red
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

# Creating Multiplots on Same Canvas

```python
# plt.subplot(nrows, ncols, plot_number)
plt.subplot(1,2,1)
plt.plot(x, y, 'r') # More on color options later
plt.subplot(1,2,2)
plt.plot(y, x, 'b');
```

# Matplotlib Object Oriented Method
 This means we will instantiate figure objects and then call methods or attributes from that object.
 create figure objects and then just call methods or attributes off of that object.Better approach dealing with a canvas that has multiple plots on it.

```python
# Create Figure (empty canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.plot(x,y)
axes.set_xlabel('X Label') # Notice the use of set_ to begin methods
axes.set_ylabel('y Label')
axes.set_title('Title')
```

add more than one axis to the figure:

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

# subplots()
The plt.subplots() object will act as a more automatic axis manager. 

```python
axes
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots(nrows=1,ncols=2)

# To fix overlaps
plt.tight_layout()

# for current_ax in axes:
#     current_ax.plot(x,y)


axes[0].plot(x, y)
axes[0].set_title('First plot')
axes[1].plot(y,x)
axes[1].set_title('Second plot')

# To fix overlaps
plt.tight_layout()
```

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title');
```

# Figure size, aspect ratio and DPI

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. It can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches
dpi is the dots-per-inch (pixel per inch).
For example:

```python
fig = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

# Legends, labels and titles

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label = 'X squared')
ax.plot(x,x**3, label = 'X cubed' )
ax.legend()
```
