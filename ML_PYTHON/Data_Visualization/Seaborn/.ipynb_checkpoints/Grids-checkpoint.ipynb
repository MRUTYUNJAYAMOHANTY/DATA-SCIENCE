# Grids
Grids are general types of plots that allow us to map plot types to rows and columns of a grid, this helps us create similar    plots separated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

# PairGrid
Pairgrid is a subplot grid for plotting pairwise relationships in a dataset.

```python
# Just the Grid
sns.PairGrid(iris)
```

```python
# map to the grid
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
# Map to upper,lower, and diagonal
g = sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# pairplot
pairplot is a simpler version of PairGrid (use quite often)

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species',palette='rainbow')
```

# Facet Grid
FacetGrid is the general way to create grids of plots based off of a feature:

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Just the Grid
g = sns.FacetGrid(tips, col="time", row="smoker")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill")
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker",hue='sex')
# Notice hwo the arguments come after plt.scatter call
g = g.map(plt.scatter, "total_bill", "tip").add_legend()
```


# JointGrid
JointGrid is the general version for jointplot() type grids, for a quick example:

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
```

```python
g = sns.JointGrid(x="total_bill", y="tip", data=tips)
g = g.plot(sns.regplot, sns.distplot)
```
